# Предсказание курса биткоина с помощью линейной регрессии

```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/aiedu-courses/stepik_applied_tasks/main/datasets/BTC_5min_sample.csv",
                   parse_dates=True, index_col='timestamp')
```

```python
data = data[['open', 'close', 'volume']]
```

```python
plt.figure(figsize=(13, 3))
plt.plot(data.open)
```

```python
day = data['open'].index.weekday
```

```python
def preprosses_data(data, lag_start=1, lag_end=10, test_size=0.15):

    # Выделяем временные лаги для цены открытия (зависимая перменная)
    # добавляем лаги для экзогенных признаков (close и volume)
    for i in range(lag_start, lag_end):
        data[f'lag_{i}_open'] = data['open'].shift(i)
        data[f'lag_{i}_close'] = data['close'].shift(i)
        data[f'lag_{i}_volume'] = data['volume'].shift(i)
        
    data = data.dropna()

    data['day'] = data.index.weekday
    data['hour'] = data.index.hour

    # Узнаем границу перехода из train в тест, чтобы агрегировать данные только из тренировочной выборки
    date_train_end = data.iloc[int(len(data)*test_size)].name
    train = data[date_train_end:]

    # Агрегируем среднее значение по дню недели
    mean_day = train.groupby('day').open.mean()
    data['mean_day'] = list(map(mean_day.get, data.day))

    # Агрегируем среднее значение по часу
    mean_hour = train.groupby('hour').open.mean()
    data['mean_hour'] = list(map(mean_day.get, data.day))

    # Удаляем столбцы часа и дня, т.к. для анализа они не нужны
    data = data.drop(['day', 'hour'], axis=1)
    
    # Выделяем зависимую переменную
    X = data.drop('open', axis=1)
    y = data['open']
                                                    
    # делим данные на тренировочные и тестовые
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, shuffle=False)
    
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test= preprosses_data(data)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
predict = pd.Series(lr.predict(X_test), index=y_test.index)
```

```python
plt.figure(figsize=(13, 3))
plt.plot(y_train, label='stock_train')
plt.plot(predict, label='prediction')
plt.plot(y_test, label='real_stock')
plt.legend()
```

```python
plt.figure(figsize=(13, 3))
plt.plot(predict, label='prediction')
plt.plot(y_test, label='real_stock')
plt.title(f'MAE: {mean_absolute_error(y_test, predict)}\nPercantages: {mean_absolute_percentage_error(predict, y_test):.4f}')
```
